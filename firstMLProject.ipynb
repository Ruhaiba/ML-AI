```python
#import the model from scikit learn library
from sklearn.linear_model import LinearRegression
```

```python
#creating linear regression object
linear_model = LinearRegression()
```

```python
# Feature variables are going to be X and Targets variable are going to be Y
X = df[['sqft']]
Y = df[['Price']]
```

```python
#model fitting
linear_model.fit(X,Y)
```

```python
print("c_0 = ", linear_model.intercept_)
print("c_1 = ", linear_model.coef_)
```

```python
linear_model.predict(np.array([[1130]]))
```
